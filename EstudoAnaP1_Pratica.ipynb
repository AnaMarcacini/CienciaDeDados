### Mode

```python
pd.Series.mode 
```

```python
carros2.head()
```

.agg(pd.Series.mode): agg() é uma função de agregação que recebe uma função como argumento. Neste caso, a função passada é pd.Series.mode, que calcula a moda (valor mais comum) de uma série do Pandas. Então, para cada grupo de 'DriveTrain', esta função retorna o valor que ocorre com maior frequência.

Resumindo, este código agrupa os dados pelo tipo de veículo ('Type') e então encontra o tipo de tração ('DriveTrain') mais comum dentro de cada grupo.


```python
carros2.groupby('Type').DriveTrain.agg( pd.Series.mode ) 
```

### Apply over row

```python
filme.apply(lambda x: x.duration/60,axis=1)
```

```python
def score_director(num_critic_for_reviews,director_facebook_likes):

  if director_facebook_likes > 0:
    return num_critic_for_reviews * director_facebook_likes
  else:
    return num_critic_for_reviews
```

```python
filme.apply(lambda row: score_director(row.num_critic_for_reviews,row.director_facebook_likes), axis=1)
```

```python
def score_director(row):

  if row.director_facebook_likes > 0:
    return row.num_critic_for_reviews * row.director_facebook_likes
  else:
    return row.num_critic_for_reviews
```

```python
filme.apply(lambda row: score_director(row), axis=1)
```

```python
filme.apply(score_director,axis=1)
```

### Com `import re` e `apply`


```python
import re

def valida_data(data):
  regex = r"^\d{4}-\d{2}-\d{2}$"
  return re.match(regex, data) is not None
```

```python
inspecao['Date'].apply(valida_data)
```

```python
inspecao[ ~ inspecao['Date'].apply(valida_data) ]
```
