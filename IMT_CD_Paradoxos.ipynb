# Correlação $\not \longrightarrow$ causalidade!

Se há uma coisa que você deve saber sobre correlação, é que correlação não é o mesmo que causalidade. Duas variáveis, como altura e notas em matemática em crianças em idade escolar, podem estar correlacionadas, embora ser bom em matemática não o torne mais alto, ou vice-versa. O que acontece é que apenas que as crianças mais velhas são mais altas e melhores em matemática. A correlação pode surgir de uma causa comum levando ambas as variáveis ​​na mesma direção.
