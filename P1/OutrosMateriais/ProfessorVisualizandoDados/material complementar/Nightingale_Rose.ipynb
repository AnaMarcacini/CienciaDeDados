## Nightingale Rose

Florence Nightingale (1820-1910, England) foi uma reformadora, estatística e é considerada a fundadora da enfermagem moderna. Ela serviu como chefe de enfermeiras durante a Guerra da Criméia organizando o atendimento aos soldados  mudando definitivamente a imagem e reputação da enfermagem. Os gráficos Nightingale Rose aparecem em seu *A Contribution to the the Sanitary History of the British Army During the Late War with Russia*, entre muitas outras visualizações de dados que produz, para demonstrar a necessidade de cuidados médicos aos soldados, muito mais vítimas de falta de tratamento e doenças do que pelo confronto armado.

Concluímos então reproduzindo gráficos Nightingale Rose a partir dos mesmos dados empregados por Nightingale para o seu célebre texto.



```python
nightingale = pd.read_csv('https://raw.githubusercontent.com/Rogerio-mack/Visualizacao-de-Dados-em-Python/main/data/nightingale.csv',index_col=0)
nightingale.head()
```

```python
f = plt.figure(figsize=(10,5))

df = nightingale[ nightingale.Year == 1855 ] 
chaves = df.Month
valores1 =  df.Disease / (df.Disease + df.Wounds)
valores2 = df.Wounds / (df.Disease + df.Wounds)

cmap2 = cm.Blues_r(np.linspace(0, 0.5, len(chaves))) 
cmap3 = cm.bwr_r(np.linspace(0.1, 0.4, len(chaves)))

ax = plt.subplot(1,2,1, polar=True)
theta = np.linspace(0.0, 2 * np.pi, len(chaves), endpoint=False)
plt.bar(theta, valores1, width=(2*np.pi) / len(chaves), color=cmap3, bottom=0, alpha=0.7, label='Disease',edgecolor='k')
plt.bar(theta, valores2, width=(2*np.pi) / len(chaves), color=cmap2, bottom=0, alpha=0.8, label='Wounds',edgecolor='k')
plt.yticks([]) # (np.linspace(0, 1, 4), color='k',fontsize=8)
plt.xticks(theta,labels=chaves,fontsize=12)
plt.title('Army Injures (Nightingale, 1854)\n', fontsize=14, weight='bold')

ax.set_theta_zero_location("N")  # theta=0 at the top
ax.set_theta_direction(-1)
plt.legend(loc='best', bbox_to_anchor=(0.5, -0.4, 0.65, 0.5))


df = nightingale[ nightingale.Year == 1854 ] 
chaves = df.Month
valores1 = df['Wounds.rate']
valores2 = df['Other.rate']
valores3 = df['Disease.rate']

ax = plt.subplot(1,2,2, polar=True)
theta = np.linspace(0.0, 2 * np.pi, len(chaves), endpoint=False)
plt.bar(theta, valores1, width=(2*np.pi) / len(chaves), color='orange', bottom=0, alpha=0.9, label='Wounds',edgecolor='k')
plt.bar(theta, valores2, width=(2*np.pi) / len(chaves), color='k', bottom=valores1, alpha=0.7, label='Other',edgecolor='k')
plt.bar(theta, valores3, width=(2*np.pi) / len(chaves), color='grey', bottom=valores1+valores2, alpha=0.5, label='Disease',edgecolor='k')

plt.yticks([]) # (np.linspace(0, 1, 4), color='k',fontsize=8)
plt.xticks(theta,labels=chaves,fontsize=12)
plt.title('Army Injures (Nightingale, 1854)\n', fontsize=14, weight='bold')

ax.set_theta_zero_location("N")  # theta=0 at the top
ax.set_theta_direction(-1)
plt.legend(loc='best', bbox_to_anchor=(0.5, -0.4, 0.65, 0.5))

plt.tight_layout()
plt.show()

```
