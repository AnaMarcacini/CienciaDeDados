# Q53. Normalização de dados

No `dataframe` da questão anterior normalize os dados de anos de escolaridade e emissões aplicando a normalização padrão (`standard`). **Qual a média e desvio padrão obtida dos valores normalizados?**.

<br>

Nota:
1. Normalização padrão (`standard`):

$$ x_{std} = \frac{x - \bar{x}}{\sigma(x)}$$

> onde $\bar{x}$ é o valor médio dos valores e $\sigma(x)$ o desvio padrão.

Dica:
2. Você pode atribuir os valores diretamente ou criar uma função para isso, empregando ou não o `apply()`.



```python
# seu código

df4[["school_years","c02"]]
```

```python
df4["school_years_norm"] = (df4["school_years"] - df4["school_years"].mean())/df4["school_years"].std()
df4.head()
```

```python
df4.describe()
```

```python
df4.school_years_norm.mean()
```

```python
df4.school_years_norm.std()
```

# Q26. Reviews $\times$ Preços, influenciam de que forma?

**Considere os imóveis com as 100 maiores quantidades de "revisões" (`number_of_reviews`) e os imóveis com as 100 menores quantidades de "revisões".  Existe diferença do preço médio dos imóveis entre esses dois grupos? Qual é essa diferença?**

<br>

Dica:
1. Atenção: Os imóveis com as 100 maiores (menores) quantidades de reviews não pode ser obtida direta e somente com `nlargest(100)` (`nsmallest(100)`) (por que???). Mesmo assim esses comandos podem ser úteis aqui.

> $[1, 2, 3, 3, 3, 4, 5, 6]$, por exemplo tem `nsmallest(3)` = $[1, 2, 3]$, mas $[1, 2, 3, 3, 3]$ ocorrências com os 3 menores valores!



```python
# seu código

df.head()
```

```python
df[["number_of_reviews"]].nsmallest(100,"number_of_reviews").mean() - df[["number_of_reviews"]].nlargest(100,"number_of_reviews").mean()
```

```python
df[["number_of_reviews","price"]].nsmallest(100,"number_of_reviews")["price"].mean() - df[["number_of_reviews","price"]].nlargest(100,"number_of_reviews")["price"].mean()
```
